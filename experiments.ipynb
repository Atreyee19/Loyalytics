```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data = pd.read_csv("/home/desktop02/LLM/DL-prereq/Churn_Modelling.csv")
data.head()
```

```python
#Preprocess the data
#Drop irrelevant columns

data.drop(columns={'RowNumber','CustomerId','Surname'},axis=1,inplace=True)
```

```python
data.head()
```

```python
#Encode Categorical Variable


#Gender
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
```

```python
#One hot Encoding Geography Column
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
```

```python
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
#Combine one hot encoded column with original data

data.drop(columns={'Geography'},axis =1)
data.head()
```

```python
data = pd.concat([data,geo_encoded_df],axis=1)
```

```python
data.head()
```

```python
#Save the encoders and standardscalers

with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
    
with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
#Divide the dataset into dependent and independent features
x = data.drop(['Exited'],axis=1)
y = data['Exited']
```

```python
#Split the dataset
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
#Scale these Features
x_train = x_train.select_dtypes(include=[float, int])
x_test = x_test.select_dtypes(include=[float, int])

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

## ANN Implementation

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
#Building the model
model = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),   ##Hidden LAyer 1 : Connected with Input Layer
    Dense(32,activation='relu'),                                   ##Hidden Layer 2
    Dense(1,activation='sigmoid')                                  ##Hidden Layer 3 : Connected with Output Layer
]    
)
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = tf.keras.losses.BinaryCrossentropy()
```

```python
#Compile the model
model.compile(optimizer=opt,loss = loss,metrics=['accuracy'])
```

```python
#Setting up the tensorboard
log_dir = "logs/fit" + datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Setting up Earlystopping
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
#Train the model
history = model.fit(
    x_train,y_train,validation_data = (x_test,y_test),epochs=100,
    callbacks = [tensorflow_callback,early_stopping_callback]
    
)
```
